```python
import pandas as pd

# Read the file you manually reviewed (never write to it here!)
matches_reviewed = pd.read_csv("subreddit_country_matches_v1.csv")

# Quick sanity check
print(matches_reviewed["status"].value_counts())
```

```python
approved_df = matches_reviewed.query("status == 'approved'")
rejected_df = matches_reviewed.query("status == 'rejected'")
unsure_df   = matches_reviewed.query("status == 'unsure'")
wrong_df    = matches_reviewed.query("status == 'wrong'")

print(f"Approved: {len(approved_df)}")
print(f"Rejected: {len(rejected_df)}")
print(f"Unsure: {len(unsure_df)}")
print(f"Wrong: {len(wrong_df)}")
```

```python
approved_df.to_csv("subreddit_matches_approved.csv", index=False)
rejected_df.to_csv("subreddit_matches_rejected.csv", index=False)
unsure_df.to_csv("subreddit_matches_unsure.csv", index=False)
wrong_df.to_csv("subreddit_matches_wrong.csv", index=False)
```

```python
import pandas as pd

approved_df = pd.read_csv("subreddit_matches_approved.csv")
rejected_df = pd.read_csv("subreddit_matches_rejected.csv")
unsure_df   = pd.read_csv("subreddit_matches_unsure.csv")
group_df    = pd.read_csv("subreddit_matches_group.csv")
```
